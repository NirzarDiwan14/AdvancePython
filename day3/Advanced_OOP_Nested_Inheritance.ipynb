## 1. Nested Classes

A **nested class** is a class defined inside another class.

### Why companies use nested classes:
- Logical grouping
- Encapsulation of helper logic
- Cleaner namespaces

### Real-world examples:
- Address inside User
- Config inside Service
- Response models inside APIs

```python

class Company:
    class Employee:
        def __init__(self, name):
            self.name = name

        def show(self):
            print(f"Employee Name: {self.name}")

emp = Company.Employee("Amit")
emp.show()
```

## 2. Instance Methods

Instance methods:
- Access object state
- Modify instance variables
- Core business logic

```python

class Order:
    def __init__(self, amount):
        self.amount = amount

    def apply_discount(self, percent):
        self.amount -= self.amount * percent / 100

order = Order(2000)
order.apply_discount(10)
print(order.amount)
```

```python
# Your code here
class Wallet:
    __money = 0
    def __init__(self,owner):
        self.owner = owner
    def show_balance(self):
        print(self.__money)
    
    def add_money(self,amount):
        self.__money += amount
    def spend_money(self,amount):
        if amount <= self.__money:
            self.__money -= amount
        else:
            print("Insufficient Balance")


w1 = Wallet("Nirzar")
w1.show_balance()
w1.add_money(1000)
w1.show_balance()
# w1.spend_money(1500) 
w1.spend_money(450)
w1.show_balance()



        
```

## 3. Class Methods

Class methods operate on class-level data.

### Industry use:
- Factory patterns
- Global configuration

```python

class Application:
    environment = "DEV"

    @classmethod
    def set_environment(cls, env):
        cls.environment = env

Application.set_environment("PROD")
print(Application.environment)
```

### ðŸ§ª Task 3
Create a class `Logger` with class variable `level` and class method to update it.

```python
# Your code here
class Logger:
    level = 0
    @classmethod
    def set_level(cls,lvl):
        cls.level = lvl
    
Logger.set_level(5)
print(Logger.level)
```

## 4. Static Methods

Static methods:
- Utility logic
- No object or class state access

```python

class Validator:
    @staticmethod
    def is_positive(num):
        return num > 0

print(Validator.is_positive(10))
```

## 5. Inheritance

Inheritance enables code reuse and hierarchical design.

### Used heavily in frameworks (Django, FastAPI)

```python

class User:
    def login(self):
        print("User logged in")

class Admin(User):
    def delete_user(self):
        print("User deleted")

admin = Admin()
admin.login()
admin.delete_user()
```

## 6. super() and Constructor Chaining

`super()` calls the parent class constructor.

### Mandatory in real projects

```python

class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id

emp = Employee("Ravi", 101)
print(emp.name, emp.emp_id)
```

## 7. Method Overriding

Child class provides its own implementation.

```python

class Notification:
    def send(self):
        print("Sending notification")

class EmailNotification(Notification):
    def send(self):
        print("Sending Email")

n = EmailNotification()
n.send()
```

## 8. Method Overloading (Python Style)

Python does NOT support traditional overloading.
We simulate using:
- Default arguments
- *args

```python

class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(2))
print(calc.add(2, 3))
print(calc.add(2, 3, 4))
```
