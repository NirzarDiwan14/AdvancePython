
## 4Ô∏è‚É£ __name__ == "__main__"

### Why It Exists
- Separates script execution from module usage
- Enables reusability & testing

### Industry Use
- Entry points
- CLI tools



## üìù Task 4

Why is __name__ == "__main__" important?


```python
# Your explanation here
when we use one or more modules then at the time of importing the the file the default python behaviour runs all the code of imported modules for one time and then cache it , sometimes it makes the problem by running some code which was not intended so to tackle that problem we use name == main method , by doing that now when we import the file then opnly the part which is outside the name == main will be getting excecuted and we will avoid errors 
```

```python
# Review of your Task 4 answer:

# ‚úÖ CORRECT UNDERSTANDING:
# You identified the core issue: unintended code execution on import
# You mentioned caching (Python's module caching is real)

# ‚ö° KEY POINTS TO REFINE:

# 1. CLEARER EXPLANATION:
# When you import a module, Python executes ALL top-level code
# __name__ == "__main__" creates a "gate" for code that should ONLY run when:
#    - The file is executed directly (not imported)

# 2. PRACTICAL EXAMPLE:
# Without __name__ == "__main__":
def process_data():
    print("Processing...")

process_data()  # ‚ùå RUNS EVERY TIME THIS MODULE IS IMPORTED!

# With __name__ == "__main__":
def process_data():
    print("Processing...")

if __name__ == "__main__":
    process_data()  # ‚úÖ ONLY RUNS when file is executed directly

# 3. WHY IT MATTERS FOR MODULARITY:
# - Modules should export functions/classes, NOT execute logic
# - Entry points (main.py) execute logic using imported functions
# - This separation enables reusability & testing

# YOUR NEXT STEP:
# Move to Task 5 - write pseudo-code for dependency injection
```

```python
# ‚ùå BAD: Hard-coded dependency
class PaymentService:
    def __init__(self):
        self.db = Database()  # Tightly coupled!
    
    def process_payment(self, user_id, amount):
        # Service creates its own database
        return self.db.save_transaction(user_id, amount)

# Usage
service = PaymentService()
service.process_payment(1, 100)

# Problem: Can't test without real database, can't swap database, inflexible


# ‚úÖ GOOD: Dependency Injection
class PaymentService:
    def __init__(self, db):  # Database passed in
        self.db = db
    
    def process_payment(self, user_id, amount):
        return self.db.save_transaction(user_id, amount)

# Usage
db = Database()
service = PaymentService(db)
service.process_payment(1, 100)

# Testing with mock:
mock_db = MockDatabase()
service = PaymentService(mock_db)  # Easy to swap!
service.process_payment(1, 100)
```
