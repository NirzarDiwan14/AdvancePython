## 1Ô∏è‚É£ What is Regex?

Regular Expressions are patterns used to **search, match, and manipulate text**.

We use Python's built-in `re` module.

```python
import re

text = "Hello World! My phone number is 123-456-7890."
```

## 2Ô∏è‚É£ Basic Pattern Matching

Let's search for a word inside text.

```python

pattern = re.compile(r"phone")
matches = pattern.finditer(text)

for match in matches:
    print(match.group())
```

## 3Ô∏è‚É£ Matching Digits

`\d` ‚Üí matches digits (0‚Äì9)

```python

text = "Order numbers: 123 456 789"
pattern = re.compile(r"\d{3,4}")

for match in pattern.finditer(text):
    print(match.group())
```

## 4Ô∏è‚É£ Matching Phone Numbers

```python

text = """
123-456-7890
987-654-3210
123.456.7890
"""

pattern = re.compile(r"\d{3}[-]\d{3}[-]\d{4}")

for match in pattern.finditer(text):
    print(match.group())
```

## 5Ô∏è‚É£ Matching Names with Titles

```python

text = """
Mr. Smith
Mrs. Robinson
Ms Davis
Mr T
"""

pattern = re.compile(r"Mr\.?\s[a-zA-Z]+")

for match in pattern.finditer(text):
    print(match.group())
```

## 6Ô∏è‚É£ findall vs finditer

```python

text = "cat bat rat mat"

print(re.findall(r"\w+at", text))
```

## 7Ô∏è‚É£ Email Matching

```python

text = """
test@gmail.com
user_123@yahoo.in
user1@gmail.co.in
invalid@email
"""

pattern = re.compile(r"[a-zA-Z0-9_.]+@[a-zA-Z]+\.[a-zA-Z]+\.[a-zA-Z]+")

for match in pattern.finditer(text):
    print(match.group())
```
